```python
def clean_data(data):
    raw_rating=0
    text_to_num={"one":1,"two":2,"three":3,"four":4,"five":5}
    cleaned_data=[]
    unique_names=set()
    
    for item in data:
       if isinstance(item["rating"],str):
           raw_rating=str(item["rating"].strip().lower())
       if raw_rating in text_to_num:
                raw_rating=text_to_num[raw_rating]
                item["rating"]=raw_rating
        
       raw_age=item.get("age")
       if raw_age==None:
                item["age"]=None
       if item["name"].strip() in unique_names:
           continue
      
        item["name"]=item["name"].strip()
        unique_names.add(item["name"].strip())
        cleaned_data.append(item)item["name"]
    return cleaned_data
```

```python
data2=clean_data(data)
for i in data:
    nam=i["name"].strip
    print(nam)
```
